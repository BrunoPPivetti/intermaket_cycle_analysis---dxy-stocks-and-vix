```python
from fredapi import Fred
import pandas as pd
import yfinance as yf
import os 
from dotenv import load_dotenv
```

```python
fred_api_key = os.getenv('API_FRED')
fred = Fred(api_key=fred_api_key)
```

```python
#Downloading 2y T-Note
get_t_note_2y = fred.get_series('DGS2')
```

```python
#Getting it into a dataframe
t_note_2y = pd.DataFrame(get_t_note_2y, columns=['Close'])
t_note_2y.index.name = 'Date'
t_note_2y.head()

t_note_2y.to_csv('US02Y.csv')
```

```python
#Downloading S&P500 
get_sp500 = yf.download('^GSPC', start='1985-01-01')
```

```python
#Cleaning data and getting into a dataframe
sp500 = pd.DataFrame(get_sp500)
sp500 = sp500.droplevel('Ticker',axis=1)
sp500 = sp500.drop(['Volume'],axis=1)
sp500.to_csv('SP500.csv')
sp500.head()
```

```python
#Downloading DXY
get_dxy = yf.download('DX-Y.NYB', start='1970-01-01')
```

```python
#Cleaning data and getting into a dataframe
dxy = pd.DataFrame(get_dxy)
dxy = dxy.droplevel('Ticker',axis=1)
dxy = dxy.drop(['Volume'],axis=1)
dxy.to_csv('DXY.csv')
dxy.head()
```
